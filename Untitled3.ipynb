```python
import os
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
def preprocess_image(img_path, target_size=(640, 640)):
    img = cv2.imread(img_path)
    img = cv2.resize(img, target_size)
    img = img / 255.0  # normalize to [0,1]
    return img

# Example usage:
img = preprocess_image('/content/drive/MyDrive/YoloV8 Vollybal/valid/images/006991516a952bae_jpg.rf.fb9035e986fb5541d85ad7711248446c.jpg')
plt.imshow(cv2.cvtColor((img*255).astype('uint8'), cv2.COLOR_BGR2RGB))
plt.title("Preprocessed Image")
plt.show()
```

```python
%%writefile volleyball_data.yaml
train: /content/drive/MyDrive/YoloV8 Vollybal/train
val: /content/drive/MyDrive/YoloV8 Vollybal/valid
test: /content/drive/MyDrive/YoloV8 Vollybal/test
nc: 1  # number of classes (volleyball)
names: ['volleyball']
```

```python
import cv2
import os

# Example: resize all training images to 640x640 (skip if already done)
train_img_dir = '/content/drive/MyDrive/YoloV8 Vollybal/test/images'
resized_dir = '/content/drive/MyDrive/YoloV8 Vollybal/test/images_resized'
os.makedirs(resized_dir, exist_ok=True)

for img_name in os.listdir(train_img_dir):
    img_path = os.path.join(train_img_dir, img_name)
    img = cv2.imread(img_path)
    if img is not None:
        resized = cv2.resize(img, (640, 640))
        cv2.imwrite(os.path.join(resized_dir, img_name), resized)
```

```python
from ultralytics import YOLO

# Load a pretrained YOLOv8 model (e.g., YOLOv8s)
model = YOLO('yolov8s.pt')

# Train the model
# Adjust 'batch' based on your GPU memory and choose your learning rate (lr0)
results = model.train(data='volleyball_data.yaml', epochs=50, batch=16, lr0=0.01)
```

```python
# Evaluate the model on the validation dataset
metrics = model.val()
print(metrics)
```

```python
import time

start_time = time.time()
_ = model.predict(source='/content/drive/MyDrive/YoloV8 Vollybal/valid/images', imgsz=640)
inference_time = time.time() - start_time
print(f"Inference time on validation set: {inference_time:.2f} seconds")
```
